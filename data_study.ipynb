## Great Expectations Data Validation

```python
pged = gx.read_excel("KPMG_VI_New_raw_data_update_final.xlsx",sheet_name="Transactions",skiprows=1)
pged.head()
```

### Validation rules - Unique IDs & Data Types Validation

```python
pged.info()
```

```python
pged.expect_column_values_to_be_unique("transaction_id")
```

```python
pged.expect_column_values_to_be_of_type(column="transaction_date", type_="datetime64")
```

```python
pged.expect_column_values_to_not_be_null(column=['order_status','brand'])
# One cycle's brand is not available, can be ignored
```

```python
pged.expect_column_values_to_be_between(column="list_price",min_value =10 ,max_value=2100)
```

```python
pged["profit"] = pged["list_price"] - pged["standard_cost"] 
pged
```

## Plotly graphs

```python

# fig = px.line(pged, x='brand', y='transaction_id')
fig = px.histogram(pged, x="brand", y="transaction_id",
             labels=dict(x = "number of customers"),
             height=500,
              title = "Number of sales by each brand",
              histfunc = "count"
             )
# Show the chart
fig.show()
```
